```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import random

lst = [1,51,101,151,201,251]
year_url = ["2000-01-01","2018-12-30"]
movie_name = []
movie_year = []
movie_rating = []
movie_metascore = []
movie_vote_number = []

for i in lst:
    url = ("https://www.imdb.com/search/title/?title_type=feature&release_date={},{}&start={}&ref_=adv_nxt".format(year_url[0],year_url[1],i))
    response = requests.get(url)
    time.sleep(random.randint(5,8))
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')
    liste_films = soup.find_all(class_="lister-list")
    liste_films = liste_films[0]
    films = liste_films.find_all("h3")
    for i in films:
        nomfilm = i.find("a").text
        movie_name.append(nomfilm)
        anneefilm = i.find(class_="lister-item-year text-muted unbold").text
        movie_year.append(anneefilm)

    #print(movie_name)
    #print(movie_year)

    films = liste_films.find_all(class_="ratings-bar")

    for i in films:
        ratingfilm = i.find("div")
        ratingfilm = ratingfilm["data-value"]
        movie_rating.append(ratingfilm)

    #print(movie_rating)

    for i in films: 
        try:
            liste_i = list(i)
            liste_i = liste_i[5]
            liste_i = list(liste_i)
            liste_i = liste_i[1].text
            movie_metascore.append(liste_i)
        except IndexError :
            movie_metascore.append("0")

    #print(movie_metascore)

    films = liste_films.find_all(class_="sort-num_votes-visible")

    for i in films:
        try:
            liste_i = list(i)
            liste_i = liste_i[3].text
            movie_vote_number.append(liste_i)
        except IndexError :
            movie_vote_number.append("0")

    #print(movie_metascore)
    df = pd.DataFrame({"Film" : movie_name,
                   "Ann√©e" : movie_year,
                   "Rating" : movie_rating,
                  "Meta-Score" : movie_metascore,
                  "Vote Number" : movie_vote_number})
    df.to_csv("Best_Movies_imbd.csv")
```
