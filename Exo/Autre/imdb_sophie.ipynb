```python
# modules importés
import requests
from bs4 import BeautifulSoup
import time 
from random import randint


# Listes vides
title=[]
year=[]
rate_imdb=[]
meta=[]
votes_numb=[]

# demander le nombre de pages
start=[1,51]

# demander les années 
#year_debut=int(input("A partir de quelle année (yyyy)?"))
#year_fin=int(input("Jusqu'à quelle année (yyyy)?"))
#year_url=year_debut+"-01-01"+","+year_fin+"-12-31"


# sortir les urls 
for i in start:
    for year_url in range(2000,2001):
        url=f"https://www.imdb.com/search/title/?title_type=feature&release_date={year_url}&user_rating=5.0,9.8&sort=num_votes,desc&start={i}&ref_=adv_nxt"
        print(url)
        get=requests.get(url)
        # marquer 8 à 15 secondes de pause entre chaque requête
        time.sleep(randint(8,15))
        # récupérer le code de statut de la requête et renvoyer si besoin une erreur
        status_code=get.status_code
        if status_code!=200:
            print("Oups, something went wrong")
            break

        # parcourir le code html
        html=get.content
        soup=BeautifulSoup(html, "html.parser")

        # Sélectionner les 50 films de chaque page
        # Écrire une boucle qui parcourt tout les containers
        for containers in soup.select(".ratings-metascore span"):
            #print(containers)
            if containers.select(".lister-item mode-advanced .ratings-metascore") is not None : # si le container contient un métascore
                print(containers.text);
                #print("ok")

                # Extraire le titre & append la liste title
                for j in soup.select(".lister-item-header a"):
                    #print(j.text)
                    title.append(j.text)

                # Extraire l'année & append la liste year
                for k in soup.find_all(class_="lister-item-year text-muted unbold"):
                    #print(k.text)
                    year.append(k.text)

                # Extraire la note imdb & append la liste rate_imdb
                for l in soup.find_all(class_="inline-block ratings-imdb-rating"):
                    #print(l.text)
                    rate_imdb.append(l.text)

                # Extraire le metascore & append la liste meta
                for m in soup.select(".ratings-metascore span"):
                    #print(m.text)
                    if m.text=="":
                        meta.append("0")
                    else:
                        meta.append(m.text)

                # Extraire le nombre de votes avec attrs/data value 
                soup_votes=soup.find_all(class_="sort-num_votes-visible")
                for n in soup_votes:
                    #soup.find("span", text="Votes:").find_next("span")["data-value"]:
                    #print(n)
                    o=n.find("span", text="Votes:").find_next("span")["data-value"]
                    #print(o)
                    votes_numb.append(o)

                        # Gross_value = soup.find('span', text='Gross:').find_next('span').text
                        # soup.find('span', text='Gross:').find_next('span')['data-value']

            else:
                print("métascore indisponible, nous n'avons pas extrait les données")
```

```python
print(len(title))
print(len(year))
print(len(rate_imdb))
print(len(meta))
print(len(votes_numb))
```
